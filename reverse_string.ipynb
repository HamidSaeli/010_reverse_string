# Project 10: Reverse a String


```python
# This project teaches string slicing: text[::-1]
```

```python
# Step 1: Ask the user to enter a string
user_text = input("Enter a text: ")
```

```python
# Step 2: Reverse the string using slicing
reversed_text = user_text[::-1]
```

```python
# Step 3: Print the reversed string
print(f"The reversed text is: {reversed_text}")
```
